# MLE and for Poisson-spike data

Run the below cells to generate some fake data!

```python
import scipy.stats as stat
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam = np.random.choice(range(20))
n = 10000

data = stat.poisson.rvs(mu=lam,size=(n,))

plt.plot(data)
plt.show()
```

You should probably have noticed that the MAP estimate is very much biased by the prior that we choose. So, how do we know which model of the prior is the best one? Well, we can just use Bayes' Rule again!

After seeing some data, let's decide which model best fits. We can directly compare the **evidence** for a model given by
\begin{equation*}
p(\mathcal{D} \mid \mathcal{M}) = \int p(\mathcal{D} \mid \theta_M) p(\theta_M \mid \mathcal{M}) d\theta_M
\end{equation*}

We can use the same principle to compare models of data from different distributions!
